```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("bares-e-restaurantes-recife.csv",sep=";")
display(df)
```

PRAÇAS

```python
df = pd.read_csv("parquespracas.csv",sep=";")
df=df.drop(columns=['nome_equip_urbano','codigo_logradouro','lei_equip_urbano','area','perimetro','codigo_bairro'])

# Supondo que df seja o seu DataFrame
df.rename(columns={'tipo_equip_urbano': 'Descricao', 'endereco_equip_urbano': 'Endereco', 'nome_oficial_equip_urbano': 'Nome','nome_bairro':'Bairro'}, inplace=True)
# Remover linhas com valores NaN na coluna 'nome'
df= df.dropna(subset=['Nome'])
df = df[['Nome', 'Descricao', 'Bairro', 'Endereco','latitude','longitude']]
display(df)
```

```python
contagem_nan = df['Nome'].isna().sum()
print(contagem_nan)
```

MUSEUS

```python
df1 = pd.read_csv("museu-1.csv",sep=";")
df1=df1.drop(columns=['site','telefone'])

# Supondo que df seja o seu DataFrame
df1.rename(columns={'nome': 'Nome', 'descricao': 'Descricao', 'bairro': 'Bairro','logradouro':'Endereco',}, inplace=True)

display(df1)
```

TEATROS

```python
df2 = pd.read_csv("teatros.csv",sep=";")
df2=df2.drop(columns=['Telefone'])

# Supondo que df seja o seu DataFrame
df2.rename(columns={'nome': 'Nome', 'descricao': 'Descricao', 'bairro': 'Bairro','logradouro':'Endereco',}, inplace=True)

display(df2)
```

MERCADOSPUBLICOS

```python
df3 = pd.read_csv("mercadospublicos.csv",sep=";")
#df3=df3.drop(columns=['Telefone'])

# Supondo que df seja o seu DataFrame
df3.rename(columns={'nome': 'Nome', 'descricao': 'Descricao', 'bairro': 'Bairro','logradouro':'Endereco',}, inplace=True)

display(df3)
```

CONCAT

```python
df_concatenado = pd.concat([df, df1, df2, df3])

df_concatenado.reset_index(drop=True, inplace=True)

display(df_concatenado)
df_concatenado.to_csv('pontosTuristicos.csv', index=True)
```
